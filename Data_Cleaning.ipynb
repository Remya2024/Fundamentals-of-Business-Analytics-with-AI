```python
# prompt: Extract the data  in following link:
# https://github.com/RSRemya/Feature_Engineering/blob/main/weather_dataset.csv

import pandas as pd
weather_df = pd.read_csv('https://raw.githubusercontent.com/RSRemya/Feature_Engineering/main/weather_dataset.csv')
weather_df
```

```python
# prompt: Check the datatype of each column

weather_df.dtypes
```

```python
# prompt: Change the datatype of column : day to date and time type and then show datatypes of all columns

import pandas as pd
weather_df['day'] = pd.to_datetime(weather_df['day'])
weather_df.dtypes
```

```python
# prompt: Now make day column to be index column and show the dataset

weather_df = weather_df.set_index('day')
weather_df
```

```python
# prompt: Remove rows having only nulls values and show the dataset

weather_df.dropna(how='all', inplace=True)
weather_df
```

```python
# prompt: Conduct interpolation through method=time to deal with missing values in the data

weather_df_interpolated = weather_df.interpolate(method='time')
weather_df_interpolated
```

```python
# prompt: Fill nulls in column event with mode of the column

weather_df_interpolated['event'] = weather_df_interpolated['event'].fillna(weather_df_interpolated['event'].mode()[0])
weather_df_interpolated
```

```python
# prompt: please create a code to show only two numbers after decimal point for column temperature

weather_df_interpolated['temperature'] = weather_df_interpolated['temperature'].round(2)
weather_df_interpolated
```

```python
# prompt: Extract the data in following link:
# https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv

import pandas as pd
titanic_data = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
titanic_data
```

```python
# prompt: Find number of null values in each column

titanic_data.isnull().sum()
```

```python
# prompt: tell me the percentage of data missing in each column

# Calculate the percentage of missing values in each column
missing_percentage = (titanic_data.isnull().sum() / len(titanic_data)) * 100

# Print the result
print(missing_percentage)
```

```python
# prompt: find the number of rows in the data

num_rows = titanic_data.shape[0]
print("Number of rows:", num_rows)
```

```python
# prompt: Remove the column : Cabin

titanic_data = titanic_data.drop('Cabin', axis=1)
titanic_data
```

```python
# prompt: find the datatype of all columns

print(titanic_data.dtypes)
```

```python
# prompt: Fill missing values in column Embarked with mode value of it

titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)
```

```python
# prompt: Create boxplot for column Age

import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x=titanic_data['Age'])
plt.show()
```

```python
# prompt: Fill missing values in column Age with its median

titanic_data['Age'].fillna(titanic_data['Age'].median(), inplace=True)
```

```python
# prompt: Remove outliers in all columns in dataset by replacing them with maximum and minimum values

for column in titanic_data.select_dtypes(include=['number']).columns:
  # Calculate quartiles
  Q1 = titanic_data[column].quantile(0.25)
  Q3 = titanic_data[column].quantile(0.75)
  IQR = Q3 - Q1

  # Define upper and lower bounds for outliers
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  # Replace outliers with min and max values within bounds
  titanic_data[column] = titanic_data[column].clip(lower_bound, upper_bound)
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x=titanic_data['Age'])
plt.show()
```

```python
# prompt: Create an age group column by using binning

import pandas as pd
# Define age bins
bins = [0, 18, 30, 50, 80]
labels = ['Children', 'Young Adults', 'Adults', 'Seniors']

# Create a new column 'AgeGroup' based on the bins
titanic_data['AgeGroup'] = pd.cut(titanic_data['Age'], bins=bins, labels=labels)

# Display the updated DataFrame
titanic_data.head()
```
