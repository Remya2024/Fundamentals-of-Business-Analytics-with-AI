```python
# prompt: Extract data in the following link:
# https://www.basketball-reference.com/leagues/NBA_2015_totals.html

!pip install beautifulsoup4
!pip install lxml

import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.basketball-reference.com/leagues/NBA_2015_totals.html'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'lxml')

table = soup.find('table', id='totals_stats')
headers = [th.text for th in table.find('thead').find_all('th')]
rows = table.find('tbody').find_all('tr')

data = []
for row in rows:
  cells = row.find_all('td')
  if cells:
    row_data = [cell.text for cell in cells]
    data.append(row_data)

df = pd.DataFrame(data, columns=headers[1:])
df.head()
```

```python
# prompt: Extract data from iris dataset of sklearn library and show it as a complete dataset with proper column header and target column

from sklearn import datasets
import pandas as pd

iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
print(df)
```
